```python
def gradient(x):
    return (x+3)**2
def costfunction(x):
    return 2*(x+3)
x=2.0
learning_rate=0.1
no_of_iteration=100
x_values=[]
y_values=[]

for i in range(no_of_iteration):
    x_values.append(x)
    y_values.append(costfunction(x))
    grad=gradient(x)
    x=x-learning_rate*grad
    print(f"iteration{i+1} x:{x} and cost:{costfunction(x)}")
print(f"local minima x={x} and cost={costfunction(x)}")
    
    
```

```python
import matplotlib.pyplot as plt
plt.plot(x_values,y_values,'ro-')
plt.title('gradient descent')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
